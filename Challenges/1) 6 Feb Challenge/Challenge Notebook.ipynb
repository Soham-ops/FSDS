# Q2) Create Table

```python
cursor.execute("""create table cardataset.car(buying VARCHAR(10), maint VARCHAR(10), doors VARCHAR(10), persons VARCHAR(10),
                lug_boot VARCHAR(10), safety VARCHAR(10),labels VARCHAR(10))""")
```

```python
cursor.execute("use cardataset")
cursor.execute("show tables")
cursor.fetchall()
```

# Q3) Drop

```python
cursor.execute("drop table cardataset.car")
mydb.commit()
cursor.execute("use cardataset")
cursor.execute("show tables")
cursor.fetchall()
```

# Q4) Insert

```python
with open('car.data') as f:
    c=0
    for  i in f:
        lofs = i.split(",")
        lofs[6] = lofs[6][0:5]
        print(lofs)
        print(f"""insert into cardataset.car values('{lofs[0]}','{lofs[1]}','{lofs[2]}','{lofs[3]}','{lofs[4]}','{lofs[5]}','{lofs[6]}')""")
        cursor.execute(f"""insert into cardataset.car values('{lofs[0]}','{lofs[1]}','{lofs[2]}','{lofs[3]}','{lofs[4]}','{lofs[5]}','{lofs[6]}')""")
        c+=1
print(c)
mydb.commit()
```

# Q5) Check all records inserted or not

<li>In the above LOC the variable c is telling the number of records inserted (1728), which on comparing with the original dataset we find that number of rows is  same as c.Thus we conclude all records have been inserted</li>

# Q6)Group by

```python
cursor.execute("select * from cardataset.car")
cursor.fetchall()
```

```python
cursor.execute("select buying,count(buying) from cardataset.car group by buying")
cursor.fetchall()
```

# Q7) Where

```python
cursor.execute("select * from cardataset.car where doors='4'")
cursor.fetchall()
```
