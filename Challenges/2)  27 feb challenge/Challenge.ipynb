```python
import pandas as pd
import requests
```

# Fetching the data and converting to json

```python
url = "https://api.github.com/repos/pandas-dev/pandas/issues"
```

```python
data  = requests.get(url)
```

```python
data1 = data.json()
```

```python
data1
```

## taking all key,value pairs within user key and making them as key,value pairs at the same lvl as user key

```python
size = len(data1)
for i in range(size):
    for j,k in data1[i]['user'].items():
        data1[i]['user_'+j]=k
```

```python
data1
```

```python
#getting the key name so as to extract column in the next Line of Code
data1[0].keys()
```

# Saving obtained df into csv file

```python
df5 = pd.DataFrame(data1,columns = ['id','url','repository_url','labels_url','comments_url','events_url','html_url','user',
                                    'user_login', 'user_id', 'user_node_id', 'user_avatar_url', 'user_gravatar_id', 
                                    'user_url', 'user_html_url', 'user_followers_url', 'user_following_url', 'user_gists_url', 
                                    'user_starred_url', 'user_subscriptions_url', 'user_organizations_url', 
                                    'user_repos_url', 'user_events_url', 'user_received_events_url', 'user_type', 'user_site_admin'])
```
